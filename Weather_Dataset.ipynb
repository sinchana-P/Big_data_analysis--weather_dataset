What is the str contains method in Python?

Python string __contains__() is an instance method and returns boolean value True or False 

depending on whether the string object contains the specified string object or not. 

Note that the Python string contains() method is case sensitive.

*Syntax:*

# Series.str.contains(pat, case=True, flags=0, na=nan, regex=True)

Parameter : 

pat : Character sequence or regular expression. 

case : If True, case sensitive. 

flags : Flags to pass through to the re module, e.g. re.IGNORECASE. 

na : Fill value for missing values.

regex : If True, assumes the pat is a regular expression.

Returns : Series or Index of boolean values 
 

```python
# str.contains
data[ data['Weather Condition'].str.contains('Snow') ]
```

```python
data[ data['Weather Condition'].str.contains('Snow') ].head(50)
```

```python
data[ data['Weather Condition'].str.contains('Snow') ].tail(50)
```
